## UNIVARIATE
 Univariate analysis refers to the analysis of one variable and purpose of univariate analysis is to understand the distribution of values for a single variable

```python
# 1. Histogram - sns.histplot()
#  histogram is a graphical representation of the distribution of data.
#The histogram is represented by a set of rectangles, adjacent to each
#other, where each bar represent a kind of data.
fig,ax=plt.subplots()
sns.histplot(x='White_Elo',data=chess)
print("Minimum white Elo: ",chess['White_Elo'].min())
print("Maximum white Elo: ",chess['Black_Elo'].max())
ax.set_xlim(0,8000)
ax.set_ylim(1000,4000)

plt.show()
```

```python
cols = 5
rows = 5
num_cols = chess.select_dtypes(exclude='object').columns
fig = plt.figure( figsize=(cols*5, rows*5))
for i, col in enumerate(num_cols):
#for i in num_cols:    
    ax=fig.add_subplot(rows,cols,i+1)
    
    sns.histplot(x = chess[col], ax = ax)
    
fig.tight_layout()  
plt.show()
```

```python
# 2. KDE PLOT - Kernal Density Plot - sns.kdeplot()
#A kernel density estimate (KDE) plot is a method for 
#visualizing the distribution of observations in a dataset, analogous to a 
#histogram. KDE represents the data using a continuous probability density 
#curve in one or more dimensions

sns.kdeplot(x='Black_Elo', data=chess)
```

```python
#3. RUG PLOT - sns.rugplot()
#A rug plot is a plot of data for a single quantitative variable, displayed 
#as marks along an axis
sns.rugplot(x='Result-Winner', data=chess, height=.03, color='darkgreen')
sns.histplot(x='Result-Winner', data=chess, kde=True)
```

```python
#4. BOX PLOTS - sns.boxplot()
#A box plot is a graphical rendition of statistical data based on the 
#minimum, first quartile, median, third quartile, and maximum
sns.boxplot(x=chess['White_RD'])
```

```python
cols = 5
rows = 5
num_cols = chess.select_dtypes(exclude='object').columns
fig = plt.figure(figsize= (15,9))
for i, col in enumerate(num_cols):
    
    ax=fig.add_subplot( rows, cols, i+1)
    
    sns.boxplot(x=chess[col], ax=ax)
    
plt.show()
```

```python
# 5. VIOLINE PLOT - box and kernel
#A violin plot depicts distributions of numeric data for one or more groups 
#using density curves
sns.violinplot(x=chess["PlyCount"])
```

```python
# CATEGORICAL FEATURES
chess.describe(include='object')
```

```python
# 1. COUNT PLOT
cols = 5
rows = 2
fig = plt.figure(figsize= (16,6))
all_categs = chess.select_dtypes(include='object')
cat_cols = all_categs.columns[all_categs.nunique() < 10]
for i, col in enumerate(cat_cols):
    
    ax=fig.add_subplot(rows, cols, i+1)
    
    sns.countplot(x=chess[col], ax=ax)
    
    plt.xticks(rotation=45, ha='right')
    
fig.tight_layout()  
plt.show()
```

```python
# 2. PIE CHART
df = chess['Result-Winner'].value_counts()
plt.pie(df, labels=df.index, autopct="%.0f%%");
```

## Bivariate
Bivariate analysis is stated to be an analysis of any concurrent relation between two variables or attributes.

```python
# Scatter Plot
#A scatter plot is a plot of the values of Y versus the corresponding values of
#X: Vertical axis: variable Y--usually the response variable.

chess.plot.scatter(x="White_RD",y="Black_RD")
plt.show()
```

```python
chess.plot.scatter(x="White_Elo",y="Black_Elo")
plt.show()
```

```python
# Line Plot
#A line chart or line graph or curve chart is a type of chart which displays 
#information as a series of data points called 'markers' connected by straight 
#line segments
x = chess.Black_RD
y = chess.White_RD
plt.plot(x, y)
plt.show()
```

```python
# Line Plot
x = chess.Black_Elo
y = chess.White_Elo
plt.plot(x, y)
plt.show()
```

```python
# Correlation Matrix
sns.heatmap(chess[["White_Elo","Black_Elo"]].corr(), annot=True, cmap="Blues")
```

```python
sns.heatmap(chess[["White_RD","Black_RD"]].corr(), annot=True, cmap="Blues")
```

```python
# CATEGORICAL - NUMERICAL ASSOCIATIONS
# Box Plot
#a simple way of representing statistical data on a plot in which a rectangle 
#is drawn to represent the second and third quartiles, usually with a vertical 
#line inside to indicate the median value
sns.boxplot(data=chess, x="Black_RD", y="White_RD")
plt.show()
```

```python
sns.boxplot(data=chess, x="Black_Elo", y="White_Elo")
plt.show()
```

```python
# Violin Plot
sns.violinplot(x=chess["White_Elo"], y=chess["Black_Elo"])
```
